**Are there any correlations between the history of all NBA player's height, weight, and age and points, rebounds, and assists?**

Hypothesis: Given that height and weight gives a physical advantage, this increases opportunities for players to have higher points and rebounds. For the assists, given the different type of positions and responsibilities of that position, smaller NBA players tend to have a higher assists than larger players.
